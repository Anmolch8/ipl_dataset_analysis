```python
import pandas as pd
import numpy as np
from tkinter import *
import os
from tkinter import messagebox
import datetime as dt
df=pd.read_csv('all_season_batting_card-C.csv')
#pd.set_option('max_columns', None)
df.drop(['commentary','link'],axis=1)
df2=pd.read_csv('all_season_bowling_card-C.csv')
#pd.set_option('max_columns', None)
dnf=pd.read_csv('CSE-IV- 1.csv')
def max_six():
    inter=df[['fullName','sixes']]
    batsman=inter.groupby('fullName').sum()
    batsman_sixes_dict=batsman['sixes'].to_dict()
    max_sixes=max(batsman_sixes_dict,key=batsman_sixes_dict.get)
    output1.insert(2,max_sixes)
def batsman_score(name):
    inter1=df[['fullName','runs']]
    batsman1=inter1.groupby('fullName').sum()
    batsman_runs_dict=batsman1['runs'].to_dict()
    scores= batsman_runs_dict[name]
    output.insert(0,scores)
def hundred_six():
    inter=df[['fullName','sixes']]
    batsman=inter.groupby('fullName').sum()
    batsman_sixes_dict=batsman['sixes'].to_dict()
    dda=batsman[batsman['sixes']>100]
    db=dda.reset_index()
    c=db.to_csv('D:\output.csv')
    file = "D:\output.csv"
    os.startfile(file)
def hundred_four():
    inter=df[['fullName','fours']]
    batsman=inter.groupby('fullName').sum()
    batsman_sixes_dict=batsman['fours'].to_dict()
    dda=batsman[batsman['fours']>100]
    db=dda.reset_index()
    c=db.to_csv('D:\output.csv')
    file = "D:\output.csv"
    os.startfile(file)   
def max_fours():
    inter2=df[['fullName','fours']]
    batsman2=inter2.groupby('fullName').sum()
    batsman_fours_dict=batsman2['fours'].to_dict()
    max_fours=max(batsman_fours_dict,key=batsman_fours_dict.get)
    output2.insert(2,max_fours) 
def max_strikerate():
    df['strikeRate'].replace('-', '' ,inplace=True)
    df["strikeRate"] = pd.to_numeric(df["strikeRate"])
    inter4=df[['fullName','strikeRate']]   
    batsman4=inter4.groupby('fullName').mean()
    batsman4.dropna()
    batsman_strate_dict=batsman4['strikeRate'].to_dict()
    for k, v in batsman_strate_dict.items():
        batsman_strate_dict[k] = float(v)
    batsman_strate_dict
    max_strate=max(batsman_strate_dict,key=batsman_strate_dict.get)
    output3.insert(2,max_strate)    
def matches_country():
    inter5=df[['match_id','country']]
    dd=inter5.drop_duplicates()
    inter6=dd.groupby('country').count()
    c=inter6.to_csv('D:\output.csv')
    file = "D:\output.csv"
    os.startfile(file)  
def hundred_wickets():
    inter=df2[['fullName','wickets']]
    bowler=inter.groupby('fullName').sum()
    bowler_wickets_dict=bowler['wickets'].to_dict()
    dda=bowler[bowler['wickets']>100]
    db=dda.reset_index()
    c=db.to_csv('D:\output.csv')
    file = "D:\output.csv"
    os.startfile(file) 
def max_wickets():
    inter8=df2[['fullName','wickets']]
    bowler2=inter8.groupby('fullName').sum()
    bowler_wicket_dict=bowler2['wickets'].to_dict() 
    max_wicket=max(bowler_wicket_dict,key=bowler_wicket_dict.get)
    output4.insert(2,max_wicket) 
def max_maidens():
    inter7=df2[['fullName','maidens']]
    bowler1=inter7.groupby('fullName').sum()
    bowler_madian_dict=bowler1['maidens'].to_dict()
    max_maiden=max(bowler_madian_dict,key=bowler_madian_dict.get)
    output5.insert(2,max_maiden)    
def all_stu():
    c=dnf.to_csv('D:\output.csv')
    file = "D:\output.csv"
    os.startfile(file) 
def state(name):
    dff=dnf.loc[dnf['State']==name]
    c=dff.to_csv('D:\output.csv')
    file = "D:\output.csv"
    os.startfile(file) 
def query3(state,month):
    dnf['DOB'] = pd.to_datetime(dnf['DOB'])
    dc=dnf.loc[(dnf['DOB'].dt.month==month) & (dnf['State']==state)]    
    c=dc.to_csv('D:\output.csv')
    file = "D:\output.csv"
    os.startfile(file) 
def query4(number):
    dcc=dnf.loc[dnf['Marks(CGPA)']==number]
    c=dcc.to_csv('D:\output.csv')
    file = "D:\output.csv"
    os.startfile(file) 
def query5(initial,number):
    dnf['DOB'] = pd.to_datetime(dnf['DOB'])
    dm=dnf.loc[(dnf['Name'].str.startswith(initial)) & (dnf['DOB'].dt.year==number)]
    c=dm.to_csv('D:\output.csv')
    file = "D:\output.csv"
    os.startfile(file) 
def ipl_go_back():
    toplevel.destroy()
    ipl_dataset()
def class_go_back():
    toplevel2.destroy()
    class_dataset()    
root=Tk()
root.title("Data Analysis")
root.iconbitmap("C:\\Users\\chawl\\Downloads\\app1.ico")

label0=Label(root,text='Choose the following Dataset ',anchor='center',font=('bold',20),bg='lightblue',padx=100).grid(row=0,column=1,columnspan=1)
def execute_ipl(number):
        for widgets in toplevel.winfo_children():
            widgets.destroy()
        a=0
        try:
            a=int(number)
        except ValueError:
            messagebox.showwarning("warning", " enter integer value")
            toplevel.destroy()
            ipl_dataset()
        if a==1:
            global output
            label1=Label(toplevel,text='Total scores of each batsman',font=('bold',15)).grid(row=0,column=0,sticky='w')
            name=Entry(toplevel,width=70,borderwidth=10)
            name.grid(row=2,column=0,sticky='w')
            exe=Button(toplevel,text="Enter name",borderwidth=5,command=lambda:batsman_score(str(name.get()))).grid(row=2,column=1,sticky='e')
            back=Button(toplevel,text="Go back",borderwidth=5,command=ipl_go_back).grid(row=10,column=0,sticky='e')
            output=Entry(toplevel,width=23,font=('Helvetica',24),borderwidth=4)
            output.grid(row=4,column=0,sticky='w',ipady=30,ipadx=20)
        elif a==2:
            global output1
            label1=Label(toplevel,text='Name of the batsman with maximum sixes',font=('bold',15)).grid(row=0,column=0,sticky='w')
            output1=Entry(toplevel,width=23,font=('Helvetica',24),borderwidth=4)
            back=Button(toplevel,text="Go back",borderwidth=5,command=ipl_go_back).grid(row=3,column=0,sticky='e')
            output1.grid(row=2,column=0,sticky='w',ipady=30,ipadx=20)
            max_six()
        elif a==3:
            label1=Label(toplevel,text='Batsman list who hit more than 100 sixes',font=('bold',15)).grid(row=0,column=0,sticky='w')
            btn=Button(toplevel,text="open excel sheet",borderwidth=5,command=hundred_six).grid(row=2,column=1,sticky='e')  
            back=Button(toplevel,text="Go back",borderwidth=5,command=ipl_go_back).grid(row=2,column=0,sticky='e')
        elif a==4:
            label1=Label(toplevel,text='Batsman list who hit more than 100 fours',font=('bold',15)).grid(row=0,column=0,sticky='w')
            btn=Button(toplevel,text="open excel sheet",borderwidth=5,command=hundred_four).grid(row=2,column=1,sticky='e')
            back=Button(toplevel,text="Go back",borderwidth=5,command=ipl_go_back).grid(row=2,column=0,sticky='e')
        elif a==5:
            global output2
            label1=Label(toplevel,text='Name of the batsman who hit maximun fours',font=('bold',15)).grid(row=0,column=0,sticky='w')
            output2=Entry(toplevel,width=23,font=('Helvetica',24),borderwidth=4)
            output2.grid(row=2,column=0,sticky='w',ipady=30,ipadx=20)
            back=Button(toplevel,text="Go back",borderwidth=5,command=ipl_go_back).grid(row=3,column=0,sticky='e')
            max_fours()
        elif a==6:
            global output3
            label1=Label(toplevel,text='Name of the batsman who hit maximun fours',font=('bold',15)).grid(row=0,column=0,sticky='w')
            output3=Entry(toplevel,width=23,font=('Helvetica',24),borderwidth=4)
            output3.grid(row=2,column=0,sticky='w',ipady=30,ipadx=20)
            back=Button(toplevel,text="Go back",borderwidth=5,command=ipl_go_back).grid(row=3,column=0,sticky='e')
            max_strikerate()
        elif a==7:
            label1=Label(toplevel,text='Total number of matches played in a country',font=('bold',15)).grid(row=0,column=0,sticky='w')
            btn=Button(toplevel,text="open excel sheet",borderwidth=5,command=matches_country).grid(row=2,column=1,sticky='e')
            back=Button(toplevel,text="Go back",borderwidth=5,command=ipl_go_back).grid(row=2,column=0,sticky='e')
        elif a==8:
            label1=Label(toplevel,text='Bowler who took more than 100 wickets',font=('bold',15)).grid(row=0,column=0,sticky='w')
            btn=Button(toplevel,text="open excel sheet",borderwidth=5,command=hundred_wickets).grid(row=2,column=1,sticky='e')
            back=Button(toplevel,text="Go back",borderwidth=5,command=ipl_go_back).grid(row=2,column=0,sticky='e')
        elif a==9:
            global output4
            label1=Label(toplevel,text='Bowler who took maximum wickets',font=('bold',15)).grid(row=0,column=0,sticky='w')
            output4=Entry(toplevel,width=23,font=('Helvetica',24),borderwidth=4)
            output4.grid(row=2,column=0,sticky='w',ipady=30,ipadx=20)
            back=Button(toplevel,text="Go back",borderwidth=5,command=ipl_go_back).grid(row=3,column=0,sticky='e')
            max_wickets()
        elif a==10:
            global output5
            label1=Label(toplevel,text='Bowler who thrown maximum maidens overs',font=('bold',15)).grid(row=0,column=0,sticky='w')
            output5=Entry(toplevel,width=23,font=('Helvetica',24),borderwidth=4)
            output5.grid(row=2,column=0,sticky='w',ipady=30,ipadx=20)
            back=Button(toplevel,text="Go back",borderwidth=5,command=ipl_go_back).grid(row=3,column=0,sticky='e')
            max_maidens()
        else:
            if a!=0:
                messagebox.showerror("Error", "Select right choice")
                toplevel.destroy()
                ipl_dataset()
def execute_class(number):
    for widgets in toplevel2.winfo_children():
            widgets.destroy()
    a=0
    try:
        a=int(number)
    except ValueError:
        messagebox.showwarning("warning", " enter integer value")
        toplevel2.destroy()
        class_dataset()
    if a==1:
        label1=Label(toplevel2,text='List all the students',font=('bold',15)).grid(row=0,column=0,sticky='w')
        btn=Button(toplevel2,text="open excel sheet",borderwidth=5,command=all_stu).grid(row=2,column=1,sticky='e')
        back=Button(toplevel2,text="Go back",borderwidth=5,command=class_go_back).grid(row=2,column=0,sticky='e')
    elif a==2:
        label1=Label(toplevel2,text=' Display all the students from a particular state/country',font=('bold',15)).grid(row=0,column=0,sticky='w')
        st=Entry(toplevel2,width=70,borderwidth=10)
        st.grid(row=2,column=0,sticky='w')
        exe=Button(toplevel2,text="Enter state",borderwidth=5,command=lambda:state(str(st.get()))).grid(row=2,column=1,sticky='e')
        back=Button(toplevel2,text="Go back",borderwidth=5,command=class_go_back).grid(row=5,column=1,sticky='e')
    elif a==3:
        label1=Label(toplevel2,text='Display all students from a particular state and born in a particular month',font=('bold',15)).grid(row=0,column=0,sticky='w')
        st=Entry(toplevel2,width=70,borderwidth=10)
        st.grid(row=2,column=0,sticky='w')
        st1=Entry(toplevel2,width=70,borderwidth=10)
        st1.grid(row=3,column=0,sticky='w')
        exe=Button(toplevel2,text="Enter state & month number",borderwidth=5,command=lambda:query3(str(st.get()),int(st1.get()))).grid(row=4,column=0,sticky='e')
        back=Button(toplevel2,text="Go back",borderwidth=5,command=class_go_back).grid(row=5,column=0,sticky='e')
    elif a==4:
        laabel1=Label(toplevel2,text=' Display all students with a given CGPA',font=('bold',15)).grid(row=0,column=0,sticky='w')
        st=Entry(toplevel2,width=70,borderwidth=10)
        st.grid(row=2,column=0,sticky='w')
        exe=Button(toplevel2,text="Enter CGPA",borderwidth=5,command=lambda:query4(float(st.get()))).grid(row=4,column=0,sticky='e')
        back=Button(toplevel2,text="Go back",borderwidth=5,command=class_go_back).grid(row=5,column=0,sticky='e')
    elif a==5:
        label1=Label(toplevel2,text='Display all students whose name starts with a particular letter and a particular year of birth',font=('bold',15)).grid(row=0,column=0,sticky='w')
        st=Entry(toplevel2,width=70,borderwidth=10)
        st.grid(row=2,column=0,sticky='w')
        st1=Entry(toplevel2,width=70,borderwidth=10)
        st1.grid(row=3,column=0,sticky='w')
        exe=Button(toplevel2,text="Initial & birth year",borderwidth=5,command=lambda:query5(str(st.get()),int(st1.get()))).grid(row=4,column=0,sticky='w',padx=400)
        back=Button(toplevel2,text="      Go back      ",borderwidth=5,command=class_go_back).grid(row=5,column=0,columnspan=1)
    else:
        if a!=0:
            messagebox.showerror("Error", "Select right choice")
            toplevel2.destroy()
            class_dataset()
def ipl_dataset():
        global toplevel
        toplevel=Toplevel()
        toplevel.iconbitmap('C:\\Users\\chawl\\Downloads\\app1.ico')
        lab=Label(toplevel,text='Dataset Queries',font=('bold',15),bg='lightblue',padx=270).grid(row=0,column=0)
        label1=Label(toplevel,text='1.To find Total scores of each batsman',font=('bold',15),bg='white',padx=130).grid(row=1,column=0,sticky='w')
        label2=Label(toplevel,text='2.Name of the batsman who maximum sixes ',font=('bold',15),bg='white',padx=130).grid(row=2,column=0,sticky='w')
        label3=Label(toplevel,text='3.Batsman list who hit more than 100 sixes',font=('bold',15),bg='white',padx=130).grid(row=3,column=0,sticky='w')
        label4=Label(toplevel,text='4.Batsman list who hit more than 100 fours',font=('bold',15),bg='white',padx=130).grid(row=4,column=0,sticky='w')
        label5=Label(toplevel,text='5.Name of the batsman who hit maximun fours',font=('bold',15),bg='white',padx=130).grid(row=5,column=0,sticky='w')
        label6=Label(toplevel,text='6.Batsman with maximum srtike rate    ',font=('bold',15),bg='white',padx=130).grid(row=6,column=0,sticky='w')
        label7=Label(toplevel,text='7.Total number of matches played in a country',font=('bold',15),bg='white',padx=130).grid(row=7,column=0,sticky='w')
        label8=Label(toplevel,text='8.Bowler who took more than 100 wickets',font=('bold',15),bg='white',padx=130).grid(row=8,column=0,sticky='w')
        label9=Label(toplevel,text='9.Bowler who took maximum wickets    ',font=('bold',15),bg='white',padx=130).grid(row=9,column=0,sticky='w')
        label10=Label(toplevel,text='10.Bowler who thrown maximum madien overs ',font=('bold',15),bg='white',padx=130).grid(row=10,column=0,sticky='w')
        label_ask=Label(toplevel,text='Give the Query number to execute:      ',font=('bold',15),bg='white',padx=130).grid(row=11,column=0,sticky='w')
        e=Entry(toplevel,width=50,borderwidth=5)
        e.grid(row=12,column=0,columnspan=1)
        fbtn=Button(toplevel,text="Execute",borderwidth=4,anchor='center',command=lambda:execute_ipl(e.get()),padx=50,bg='lightgreen').grid(row=13,column=0,sticky='w',padx=240)
        toplevel.geometry('600x600')
        toplevel.maxsize(600,600)

        
def class_dataset():
        global toplevel2
        toplevel2=Toplevel()
        toplevel2.iconbitmap('C:\\Users\\chawl\\Downloads\\app1.ico')
        lab=Label(toplevel2,text='Dataset Queries         ',font=('bold',15),bg='lightblue',padx=440).grid(row=0,column=0)
        label1=Label(toplevel2,text='1.List all the students                                                                                                  ',font=('bold',15),bg='white',padx=130).grid(row=1,column=0,sticky='w')
        label2=Label(toplevel2,text='2. Display all the students from a particular state/country                                             ',font=('bold',15),bg='white',padx=130).grid(row=2,column=0,sticky='w')
        label3=Label(toplevel2,text='3. Display all students from a particular state and born in a particular month                     ',font=('bold',15),bg='white',padx=130).grid(row=3,column=0,sticky='w')
        label4=Label(toplevel2,text='4. Display all students with a given CGPA                                                                           ',font=('bold',15),bg='white',padx=130).grid(row=4,column=0,sticky='w')
        label5=Label(toplevel2,text='5.Display all students whose name starts with a particular letter and a particular year of birth',font=('bold',15),bg='white',padx=130).grid(row=5,column=0,sticky='w')
        label_ask=Label(toplevel2,text='Give the Query number to execute:                                                                                 ',font=('bold',15),padx=130,bg='white').grid(row=6,column=0,sticky='w')
        e=Entry(toplevel2,width=50,borderwidth=10)
        e.grid(row=7,column=0,sticky='w',padx=120)
        fbtn=Button(toplevel2,text="Execute",borderwidth=5,command=lambda:execute_class(e.get()),background='lightgreen',padx=100).grid(row=8,column=0,sticky='w',padx=150)
        toplevel2.geometry('1500x1600')
        toplevel2.maxsize(1000,1000)

btn1=Button(root,text='IPL Dataset',anchor='center',padx=200,pady=20,command=ipl_dataset,borderwidth=4,background='lightgreen',font=('bold',10)).grid(row=1,column=1)
btn3=Button(root,text='CSE-4-1',anchor='center',padx=200,pady=20,command=class_dataset,borderwidth=4,font=('bold',10),background='lightgreen').grid(row=2,column=1)
root.geometry('550x500')
root.maxsize(550,500)
root.mainloop()
```
