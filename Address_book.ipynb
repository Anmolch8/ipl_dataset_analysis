```python
from tkinter import *
import cx_Oracle
from tkinter import messagebox
root=Tk()
root.iconbitmap("C:\\Users\\chawl\\Downloads\\address-book.ico")
root.title("Address Book")

def submit():
    con=cx_Oracle.connect("system/anmol")
    cur=con.cursor()
    try:
        cur.execute("""insert into address_book(f_name,l_name,address,city,state,zipcode,mobile_number)
        values(:f_name,:l_name,:address,:city,:state,:zipcode,:mobile_number)""",{'f_name':f_name.get(),'l_name':l_name.get(),'address':address.get(),'city':city.get(),'state':state.get(),
       'zipcode':zipcode.get(),'mobile_number':mobile_number.get()})
    except cx_Oracle.IntegrityError:
        messagebox.showinfo('Important','please enter mobile number')
    except cx_Oracle.DatabaseError:
        messegebox.showinfo('wrong value','please enter valid value')
    f_name.delete(0,END)
    l_name.delete(0,END)
    address.delete(0,END)
    city.delete(0,END)
    state.delete(0,END)
    zipcode.delete(0,END)
    mobile_number.delete(0,END)
    cur.close()
    con.commit()
    con.close()
def show():
    con=cx_Oracle.connect("system/anmol")
    cur=con.cursor()
    cur.execute('select * from address_book')
    records=cur.fetchall()
    print_record=''
    for record in records:
        print_record+=str(record)+'\n'
        label=Label(root,text=print_record,font=('arial',10)).grid(row=9,column=1,sticky='e',columnspan=2)
    cur.close()
    con.commit()
    con.close()
label=Label(root,text="Give Personal details",font=('arial',25)).grid(row=0,column=1,sticky='w')
label1=Label(root,text='NAME',font=('arial',15)).grid(row=1,column=0,sticky='w')
f_name=Entry(root,width=50)
f_name.grid(row=1,column=1,sticky='w')

label2=Label(root,text='Last name',font=('arial',15)).grid(row=2,column=0,sticky='w')
l_name=Entry(root,width=50)
l_name.grid(row=2,column=1,sticky='w')

label3=Label(root,text='Address',font=('arial',15)).grid(row=3,column=0,sticky='w')
address=Entry(root,width=50)
address.grid(row=3,column=1,sticky='w')

label4=Label(root,text='city',font=('arial',15)).grid(row=4,column=0,sticky='w')
city=Entry(root,width=50)
city.grid(row=4,column=1,sticky='w')

label5=Label(root,text='State',font=('arial',15)).grid(row=5,column=0,sticky='w')
state=Entry(root,width=50)
state.grid(row=5,column=1,sticky='w')

label6=Label(root,text='Zipcode',font=('arial',15)).grid(row=6,column=0,sticky='w')
zipcode=Entry(root,width=50)
zipcode.grid(row=6,column=1,sticky='w')

label7=Label(root,text='Mobile number',font=('arial',15)).grid(row=7,column=0,sticky='w')
mobile_number=Entry(root,width=50)
mobile_number.grid(row=7,column=1,sticky='w')
add_btn=Button(root,text='Add Record',borderwidth=5,command=submit).grid(row=8,column=1,sticky='e',pady=10)
show_btn=Button(root,text='show Records',borderwidth=5,command=show).grid(row=8,column=1,sticky='w',pady=10)

root.geometry('550x500')
root.maxsize(550,500)
root.mainloop()
```
